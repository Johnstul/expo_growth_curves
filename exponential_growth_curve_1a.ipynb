# Playing with Exponential Growth Curves:

# Import:

```python
import numpy as np

from matplotlib.pyplot import *

import matplotlib.pyplot as plt
```

# Function:

```python
def exp_gr(x,alpha):
    expo = (1 / (alpha*(np.sqrt(2*np.pi))))*(np.exp((-x**2)/2))
    return expo
```

# Implementation and Plot:

```python
factor = 3
scalar = 50
divisor = factor*10

lo = -factor*scalar
hi = factor*scalar
n = hi - lo + 1

print('lo =', lo,' hi =',hi,' n =', n)

print('\t')

lo_1 = lo/divisor
hi_1 = hi/divisor

print('lo_1 =', lo_1,' hi_1 =',hi_1,' n =', n)

x = np.linspace(lo_1,hi_1,n)

figs = 15 # figsize
figure(figsize = (figs,figs))

n_max = 15

for i in range(1,n_max):
    expo = exp_gr(x,i)
    plt.plot(expo)
#     plt.show()
```
